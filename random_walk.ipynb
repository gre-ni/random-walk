### Random Walk

Simulation and visualisation for the following problem: 

How far can you walk on a staircase, given the following rules:
- Rolling 1-2: Step 1 step backwards
- Rolling 3-5: Step 1 step forwards
- Rolling 6: Roll again and move the amount of steps on your dice


### Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Setting up parameters for experiment:
Adjust this to see different results

```python
total_rolls = 100
total_tries = 500
threshold = 80
```

### Simulation code:

```python
rng = np.random.default_rng(42)   

all_walks = []

for i in range(total_tries):
    random_walk = [0]
    for i in range(total_rolls):
        dice = rng.integers(1,7)
        step = random_walk[-1]
        
        if dice <= 2:
            step = max(0, step -1)
        elif dice > 2 and dice <= 5:
            step += 1
        else:
            step = step + rng.integers(1,7)
            
        random_walk.append(step)
    
    all_walks.append(random_walk)
```

### Visualisation of progress of the first walk

```python
example = all_walks[0]    

plt.plot(example)
plt.show()
```

### Plotting all

```python
np_aw = np.array(all_walks)
np_aw_t = np.transpose(np_aw)
```

```python
plt.plot(np_aw_t)
plt.show()
```

### Probability of reach 60+ steps:

```python
final_steps = np_aw_t[-1,:]

plt.hist(final_steps)
plt.show()
```

```python
over_sixty = len(final_steps[final_steps >= threshold])
probability = over_sixty / total_tries
print(f" Probability of reaching {threshold} steps is {round(probability * 100)} %")
```
